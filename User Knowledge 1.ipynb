```r
library(repr)
library(readxl)
library(tidyverse)
library(tidymodels)
```

```r
# Demonstrate that the dataset can be read from the web into R 
user_knowledge <- read_excel("datasets.xls", sheet = 2)
# Clean and wrangle your data into a tidy format
user_knowledge_wrangle <- user_knowledge |>
    select(STG:UNS)

# Using only training data, summarize the data in at least one table (this is exploratory data analysis). 
user_knowledge_training <- user_knowledge_wrangle |>
    select(STG, SCG, LPR, UNS) |>
    mutate(UNS = as_factor(UNS))
user_knowledge_training

user_knowledge_testing <- user_knowledge_wrangle |>
    select(STG, SCG, LPR, UNS) |>
    mutate(UNS = as_factor(UNS))
user_knowledge_testing
```

```r
user_summary <- user_knowledge_training |>
    group_by(UNS)|>
    summarize(mean_of_STG = mean(STG),
              mean_of_SCG = mean(SCG),
              mean_of_LPR = mean(LPR),
              number_of_UNS = n())
user_summary
   
```

```r
options(repr.plot.width = 10, repr.plot.height = 8)
STG_vs_SCG_plot <- user_knowledge_training|>
    ggplot(aes(x=STG, y=SCG, colour=UNS))+
    geom_point()+
    labs(x="STG(The degree of study time)", y="SCG(The degree of repetition number)", title = "The relationship between STG and SCG")+
    theme(text = element_text(size = 20))
STG_vs_SCG_plot
```

```r
options(repr.plot.width = 10, repr.plot.height = 8)
STG_vs_LPR_plot <- user_knowledge_training|>
    ggplot(aes(x=STG, y=LPR, colour=UNS))+
    geom_point()+
    labs(x="STG(The degree of study time)", y="LPR(The exam performance for related objects )", title = "The relationship between STG and LPR")+
    theme(text = element_text(size = 20))
STG_vs_LPR_plot
```
